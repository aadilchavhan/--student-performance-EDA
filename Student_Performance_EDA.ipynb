#### **Step 1: Data Collection and Loading**

```python
import pandas as pd
df = pd.read_csv ('student-mat.csv',sep=';')
df
```

#### **Step 2: Data Exploration**

```python
df.info()
```

```python
categorical_features = df.select_dtypes(include=['object']).columns
print("Categorical Features:", categorical_features)

numerical_features = df.select_dtypes(include=['number']).columns
print("Numerical Features:", numerical_features)

```

#### **Step 3: Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Distribution of student grades (G3 - final grade):

```python
plt.figure(figsize=(8,5))
sns.histplot(df['G3'],bins=15, kde=True, color='skyblue')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.title('Distribution of Final Grades')
plt.show()
```

Correlation Heatmap:

```python
numeric_df = df.select_dtypes(include=['number'])

correlation_matrix = numeric_df.corr()

plt.figure(figsize=(12, 8))


sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Student Performance (Numerical Features Only)")

plt.show()
```

Impact of Study Time on Final Grade:

```python

plt.figure(figsize=(10,5))

sns.boxplot(x=df['studytime'], y=df['G3'])

plt.xlabel("Study Time (Hours)")
plt.ylabel("Final Grade")
plt.title("Impact of Study Time on Final Grade")

plt.show()
```

Parental Education vs. Student Performance:

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['Medu'], y=df['G3'])
plt.xlabel("Parent's Education Level")
plt.ylabel('Final Grade')
plt.title("Effect of Parent's Education on Final Grades")
plt.show()
```
